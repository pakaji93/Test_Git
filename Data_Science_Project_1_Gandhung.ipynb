# Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/SkolaClass_DataScience/Week 5/student-math.csv')
df.head()
```

```python
df.info()
```

# Question 1

Nilai rata-rata akhir (G3) semua murid di sekolah “GP”

```python
Q1 = df.loc[df['school']=='GP'].agg({'G3':'mean'})
print('Nilai rata-rata ujian G3 semua siswa adalah:', Q1)
```

# Question 2

Nilai min dan max (G3) dari murid-murid di sekolah “MS”

```python
Q2 = df.loc[df['school']=='MS'].agg({'G3':['min','max']})
Q2
```

# Question 3

Proporsi jenis kelamin dari murid-murid

```python
Q3 = df['sex'].value_counts(normalize=True)
Q3
```

# Question 4

Rata-rata final grade (G3) sesuai dengan study time

```python
Q4 = df.groupby('studytime').agg({'G3':'mean'})
Q4
```

# Question 5

Rata-rata final grade (G3) sesuai dengan extra educational
support

```python
Q5 = df.groupby('schoolsup').agg({'G3':'mean'})
Q5
```

# Question 6

Rata-rata jumlah bolos (absen) sesuai dengan jumlah kegagalan

```python
Q6 = df.groupby('failures').agg({'absences':'mean'})
Q6
```

# Question 7

Jumlah anak yang meningkat skor ujiannya (first vs. final)

```python
df['growth'] = df['G3'] - df['G1']
df
```

```python
Q7 = df.loc[df['growth']>0]
Q7_J = Q7['growth'].count()
print('Jumlah anak yang meningkat skor ujiannya (first vs. final) adalah:',Q7_J)
```

# Question 8

Jumlah anak yang melakukan extra educational support (les)
sesuai free timenya

```python
Q8 = df.loc[df['schoolsup'] == 'yes'].groupby('freetime')['schoolsup'].value_counts()
Q8
```

# Question 9

Nilai rata-rata dengan bobot dari semua ujian
(first 25%, second 25%, final 50%)

```python
df['final'] = (df['G1']*1/4)+(df['G2']*1/4)+(df['G3']*1/2)
Q9 = df['final'].mean()
print('nilai rata-rata siswa dari semua ujian adalah:', Q9)
```

# Question 10

Deskripsi profil (minimal 3 fitur) dari 50 anak-anak yang
mendapatkan nilai tertinggi

```python
Q10 = df.sort_values(by='final', ascending=False)
Q10 = Q10.head(50)
Q10.reset_index(inplace=True)
Q10.rename(columns={'index':'student'}, inplace=True)
Q10
```

```python
Q10.loc[:,['student', 'G1', 'G2', 'G3', 'final', 'studytime', 'failures', 'absences']]
```

```python
med_Q10_final = Q10['final'].median()
mean_Q10_final = Q10['final'].mean()
min_Q10_studytime = Q10['studytime'].min()
max_Q10_studytime = Q10['studytime'].max()
mean_Q10_studytime = Q10['studytime'].mean()
mean_Q10_failure = Q10['failures'].mean()
max_Q10_absences = Q10['absences'].max()
Q10_desc = [{'Berikut analisa 50 siswa dengan nilai terbaik':'' ,'med_Q10_final':med_Q10_final, 'mean_Q10_final':mean_Q10_final, 'min_Q10_studytime':min_Q10_studytime, 'max_Q10_studytime':max_Q10_studytime, 'mean_Q10_studytime':mean_Q10_studytime, 'mean_Q10_failure':mean_Q10_failure, 'max_Q10_absences':max_Q10_absences}]
Q10_desc
```
